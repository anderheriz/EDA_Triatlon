# Presentación EDA

## 📖 Tema 

El proyecto analizará el rendimiento de atletas élite en triatlón comparando dos modalidades con las siguientes estructuras de carrera:

- **Distancia Olímpica:** 1.5 km natación 🏊, 40 km bicicleta 🚴 y 10 km carrera a pie🏃.
- **Distancia Ironman:** 3.8 km natación 🏊, 180 km bicicleta 🚴 y 42.2 km carrera a pie🏃.



## 📊 Datasets

### Distancia Ironman 📁

- Resultados del **Ironman World Championship Kona 2015-2025**.
    - Fuente: https://www.ironman.com/races/im-world-championship-kona/results (API pública)
    - Numero de registros: +31K filas.
        - Filtrado: Se utilizarán únicamente atletas profesionales. Los registros se reducen a casi 1K filas.

### Distancia Olímpica 📁

- Resultados de triatlón en **Juegos Olímpicos Rio 2016, Tokyo 2020/2021 y Paris 2024**  y **World Triathlon Championship Series 2016-2025** (liga de trialtón profesional).
    - Fuente:  https://events.triathlon.org/ (API pública)
    - Numero de registros: +1.3K filas.

#### Preparación de datos

Los CSVs anuales se concatenarán en dos datasets principales, uno para Ironman y otro para Olímpica. Después se homogeneizarán columnas y unidades para crear un dataframe final comparable.

Para que la comparacion sea equilibrada, el analisis final se centrara en las primeras 40 posiciones de cada carrera y genero.


## 📋 Hipótesis

1. **El segmento de bicicleta tiene mayor peso sobre el tiempo total en Ironman que en distancia olímpica.**  

2. **La carrera a pie está más relacionada con la posición final en distancia Olímpica que en Ironman.**  

3. **La variabilidad de tiempos es mayor en Ironman que en distancia Olímpica.**

4. **La diferencia porcentual de rendimiento entre hombres y mujeres varía según la modalidad y el segmento.**
